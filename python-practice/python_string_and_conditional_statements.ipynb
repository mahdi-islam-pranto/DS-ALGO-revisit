### Conditional Statements (If-Else)

Conditional statements in Python let a program make decisions. They help the program choose what to do based on certain conditions.

---

1. **`if` Statement**:
The `if` statement is the basic conditional statement.  
It runs a block of code only if the condition is true.


2. **`if–else` Statement**:
The `if–else` statement lets you run one block of code when the condition is true,  
and a different block when the condition is false.


3. **`elif` Statement**:
`elif` means "else if".  
It is used when you want to check more than one condition.  
Only the block for the first true condition will run.  
This helps keep the code clean and avoids lots of nested `if` statements.


````python
number = 40

if number > 40:
  print("passed")
else:
  print("failed")



# if -> condition :
   # operation
#  else:
    # operation
````

````python
number = 82

if number > 80:
  print("Letter Mark!")
elif number > 40:
  print("Passed")
else:
  print("Failed")
````

### Logical Operators

Logical operators are used to combine conditions.  
They help us check more than one condition at the same time.

Python has three main logical operators:

- `and`: this operator is true only if both conditions are true.  
  Example: `a > 5 and a < 10`

- `or`: this operator is true if at least one condition is true.  
  Example: `a > 5 or a == 2`

- `not`: this operator reverses the condition.  
  If something is true, `not` makes it false — and vice versa.  
  Example: `not (a > 5)`

Logical operators help make decisions in programs by combining multiple conditions.


| Operator | Description                    |
| -------- | ------------------------------ |
| `and`    | True if *both* are true        |
| `or`     | True if *at least one* is true |
| `not`    | Reverses True/False            |


````python
age = 30
salary = 100

if age>25 or salary<100:
  print("struggling")
else:
  print("Missed any one condition")


is_done = True
print(not is_done)
````

### Comparison Operators

Comparison operators in Python are used to compare two values.  
They check how one value relates to another and return True or False.

- `==` (equal to): checks if two values are the same.  
  Example: `a == b`

- `!=` (not equal to): checks if two values are different.  
  Example: `a != b`

- `>` (greater than): checks if the left value is bigger than the right value.  
  Example: `a > b`

- `<` (less than): checks if the left value is smaller than the right value.  
  Example: `a < b`

- `>=` (greater than or equal to): checks if the left value is bigger than or equal to the right value.  
  Example: `a >= b`

- `<=` (less than or equal to): checks if the left value is smaller than or equal to the right value.  
  Example: `a <= b`

These operators are commonly used in `if`, `elif`, and `while` conditions.


| Operator | Meaning          |
| -------- | ---------------- |
| `==`     | equal to         |
| `!=`     | not equal to     |
| `>`      | greater than     |
| `<`      | less than        |
| `>=`     | greater or equal |
| `<=`     | less or equal    |


````python
# = variable assign
# == Comparison operator (check equal)

age = 30
print(age == 3)
print(age != 3) # age is not equal to 3. (True)
print(age > 35)
print(age >= 35)
print(age < 35)
print(age <= 35)
````

### While Loops

In Python, a while loop keeps running as long as its condition is true. When the condition becomes false, the program moves to the next line after the loop.

A while loop in Python is used to repeat a block of code as long as a condition is true.  
When the condition becomes false, the loop stops and the program moves to the next line after the loop.

Example:  
```
cnt = 0  
while cnt < 3:  
    print("Hello")  
    cnt += 1  
```
This will print "Hello" three times.

An infinite loop happens when the condition never becomes false.  
This makes the loop run forever.

Example of an infinite loop:  
while True:  
    print("This will never stop")

To avoid infinite loops, always make sure the condition will eventually become false.
