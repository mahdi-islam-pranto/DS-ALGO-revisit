SECTION A — STRINGS

```python
# Q1
# Write a program that takes a user’s full name and:
# - Prints the name in title case
# - Prints the first and last character
# - Prints the name reversed
# - Prints how many times the letter "a" appears
# - Removes all spaces from the name

user_name = input("Enter your full name")

print(user_name.title())
print(user_name[0]+user_name[-1])
print(user_name[::-1])
print(user_name.count("a"))
print("".join(user_name.split(" ")))
# print(user_name.replace(" ", ""))
```

```python
# Q2
# Given:
# s = "Learning Python Is Fun"
# Write code to:
#  a) Convert to lowercase
#  b) Replace "Fun" with "Awesome"
#  c) Print the 3rd to 10th characters
#  d) Check if the string ends with "Fun"
#  e) Split the string into a list of words

s = "Learning Python Is Fun"
print(s.lower())
new_s = s.replace("Fun", "Awesome")
print(new_s)
print(new_s[2:9])
print(new_s.endswith("Fun"))
print(new_s.split())
```

SECTION C — IF STATEMENTS & LOGICAL OPERATORS

```python
# Q5
# Write a program that takes any number from the user and:
# - Prints Positive, Negative, or Zero
# - Prints Even or Odd
user_number = int(input("Type a number"))
if user_number == 0:
  print("Zero")
elif user_number > 0:
  print("Positive")
else:
  print("Negative")

if user_number % 2 == 0:
  print("Even")
else:
    print("Odd")
```

```python
# Q6
# Write a program that takes three subject marks from the user and:
# - Calculates total
# - Calculates average
# - If all marks ≥ 40 → print Pass, otherwise Fail
# - If average ≥ 75 → print Distinction, otherwise Normal Pass
# - Use logical operators (and, or).
first_subject = int(input("Type your first subject mark"))
second_subject = int(input("Type your second subject mark"))
third_subject = int(input("Type your third subject mark"))
print(f"-> Your total mark: {first_subject + second_subject + third_subject}")
average_mark = round((first_subject + second_subject + third_subject)/3,2)
print(f"-> Your average mark: {average_mark}")   # rounding the average

if first_subject >= 40 and second_subject >= 40 and third_subject >= 40:
  print("Pass")
else: print("Fail")

if average_mark >= 75:
  print("Distinction")
else: print("Normal Pass")
```

SECTION D — LOOPS

```python
# Q7. While Loop
# Write a program that prints numbers from 1 to 50 but:
# - Skip all numbers divisible by 4
# - Stop completely when number reaches 37

number = 1
while number <= 50:
  if number % 4 == 0:
    number += 1
    continue
  if number == 37 or number > 37:
    break
  print(number)
  number += 1
```

```python
# Q8. For Loop
# Write a program that prints the multiplication table of a user-entered number from 1 to 12.
user_number = int(input("Type a number"))
for i in range(1,13):
  print(f"{user_number} x {i} = {user_number * i}")
```

```python
# Q9. Nested Loops
# Print the following number pattern using nested loops:
# 1
# 12
# 123
# 1234
# 12345
num = 5
for i in range(1, num + 1):
  for j in range(1,i+1):
    print(j, end="")
  print()
```

SECTION E — LISTS & LIST COMPREHENSION

```python
# Q10. List Methods
# Given:
# nums = [12, 4, 8, 1, 9, 3, 4, 4]
# Write code to:
# - Sort the list
# - Insert 100 at index 2
# - Remove the value 1
# - Find and print the max and min values
# - Reverse the list
# - Count how many times 4 appears
nums = [12, 4, 8, 1, 9, 3, 4, 4]
sorted_num = sorted(nums)
print(sorted_num)
sorted_num.insert(2,100)
print(sorted_num)
sorted_num.remove(1)
print(sorted_num)
print(max(sorted_num))
print(min(sorted_num))
sorted_num.reverse()
print(sorted_num)
print(sorted_num.count(4))
```

```python
# Q11. List Processing
# Create a list of 10 integers.
#  Write a program that:
# - Prints the list
# - Prints only the odd numbers
# - Prints only the numbers greater than 20
# - Replaces all numbers less than 10 with 0

int_list = [1, 2, 7, 4, 9, 5, 12, 13, 67]
print(int_list)

odd_list = []
for i in int_list:
  if i % 2 != 0:
    odd_list.append(i)
print(odd_list)

for i in int_list:
  if i > 20:
    print(i)

for i in int_list:
  if i < 10:
    # print(int_list.index(i))
    int_list[int_list.index(i)] = 0
print(int_list)
```

```python
# Q12. List Comprehension
# Using list comprehension, create a list of all numbers from 1 to 100 that are:
# - Divisible by 3
# - But not divisible by 5

my_list = [i for i in range(1,101) if i % 3 == 0 and i % 5 != 0]

print(my_list)
```
