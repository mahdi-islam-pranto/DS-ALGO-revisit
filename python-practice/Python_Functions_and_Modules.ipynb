
## 1. **Defining Functions**

### **What is a Function?**
A reusable block of code that performs a specific task.

### Key Points
- Prevents repetition
- Organizes code
- Can take input and return output

````python
def greet():
  print("hello World")
````

````python
greet()
````

## Parameters & Arguments

### Parameter
- A variable listed in the function definition.

### Argument
- The actual value passed into the function.




````python
def greet(name): # variable name; Parameter
  print(f"hello {name}")
````

````python
greet("Sazzad") # value -> Argument
````

````python
def greet(name, age):
  print(f"hello {name}, your age is {age}")
````

````python
greet("sazzad", 80)
````

## Return Values

### Why Return?

To send a result back to where the function was called.

### Notes
- `return` ends the function
- Without return, the function returns None

````python
def add(x, y):
  result = x+y
  return result
  # print(result)

function_output = add(5, 9)
print(function_output)
````

````python
def even_odd_detector(number):
  if number%2==0:
    return "Even", True
  else:
    return "Odd", False
````

````python
even_odd_detector(7)
````

````python
s, b = even_odd_detector(7)
b
````

````python
all_tuple = (7,8,9)
x, y, z = all_tuple
y
````

## Default & Keyword Arguments

### Default Arguments
- Values that a parameter uses if the caller doesnâ€™t provide one.

### Keyword Arguments
- Arguments called using parameter names.

````python
def greet(name="User"):
  print(f"hello {name}")

greet("Fahim")
````

````python
def exp(a=1, b=3):
  return a**b

exp(2, b=5)
````

#### Avoid using mutable values (lists/dicts) as defaults.

#### Bad:
``` python
def add_item(item, lst=[]):
    lst.append(item)
```
#### Good:
``` python
def add_item(item, lst=None):
    if lst is None:
        lst = []
    lst.append(item)
```

### Variable Scope


````python
# local
# global
````

````python
x = 7

def f():
  # global x
  x = 10
  print(x-10)
  return x+10

f()
print(f())
print('done')

print(x)
````

## Try & Except
Used to handle errors gracefully.

Basic Struncture:
``` python
try:
  # some operations
except:
  # fallback

````python
try:
  int('a')
except:
  print("Something is wrong")
finally:
  print("I always run")
````

````python
try:
  int(3.9)
  int('a')
except:
  print("Something is wrong")
else:
  print("I run when try block works!")
````

````python
int('a')
````

````python
try:
    f = open("data.txt")
except FileNotFoundError:
    print("File missing!")
else:
    print("File opened!")
finally:
    print("Done!")
````
